```python
# Checking Null values in the dataset
print("Number of Null values in the dataset is",climate.isnull().sum().sum())
```

```python
# calling just those columns which have missing values more than 30 because working on them will be a mess so we will drop them later
data = pd.DataFrame(climate.isnull().sum()[climate.isnull().sum() > 30])
data
```

```python
# checking number of columns which have missing values in bulk(more than 30)
print("Number of columns which have missing values more than 30 is", data.shape)
```

```python
# Dropping the columns whoch have null values in bulk
cols_to_drop = ['F1961', 'F1962', 'F1963', 'F1964', 'F1965', 'F1966', 'F1967', 'F1968', 'F1969',
                'F1970', 'F1971', 'F1972', 'F1973', 'F1974', 'F1975', 'F1976', 'F1977', 'F1978',
                'F1979', 'F1980', 'F1981', 'F1982', 'F1983', 'F1984', 'F1985', 'F1986', 'F1987',
                'F1988', 'F1989', 'F1990', 'F1991']

climate = climate.drop(columns=cols_to_drop)
```

```python
climate.head(1)
```

```python
# Now we can see that the columns that had more than 30 missing values are dropped
```

```python
climate.isnull().sum()
```

```python
# Still, we have missing values that have to be fixed to make the data more appropriate to use
```

```python
missing = pd.DataFrame(climate.isnull().sum()[climate.isnull().sum() > 0])
```

```python
missing.shape
```

```python
# As we can see, 32 more columns have missing values
# Now we will fill the missing values
# But before filling missing values, we have to check out the datatypes of columns that have missing values
# we will fill float and int columns with a numeric value and object column's missing values with any character

# Checking datatypes of features having missing values
climate.dtypes.apply(lambda x: x.name).value_counts().reset_index().rename(columns={'index': 'DataType', 'count': 'No_of_Columns'})
```

```python
# As we observed that the numeric value has to be filled in 31+1 columns, and the character value has to be filled in 9 columns
# Fill numeric columns with a number (e.g., 0)
climate[climate.select_dtypes(include=['number']).columns] = \
    climate.select_dtypes(include=['number']).fillna(0)

# Fill object/string columns with a text (e.g., 'Unknown')
climate[climate.select_dtypes(include=['object']).columns] = \
    climate.select_dtypes(include=['object']).fillna('Unknown')
```

```python
# Let's check if our data is immaculate by missing values
print("Number of missing values after cleaning and filling is", climate.isnull().sum().sum())
```

```python
# Preparartion of data for visualization

sns.set_theme(style="whitegrid", context="talk", palette="coolwarm")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.labelsize"] = 13
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["figure.facecolor"] = "white"

# Extract year columns (numeric data)
year_cols = [col for col in climate.columns if col.startswith("F")]
climate_years = climate[year_cols]
```

```python
# Global Average Temperature Trend (1992‚Äì2022)
global_trend = climate_years.mean()
trend_climate = pd.DataFrame({"Year": [int(y[1:]) for y in year_cols], "Mean_Temp": global_trend.values})

plt.figure()
sns.lineplot(data=trend_climate, x="Year", y="Mean_Temp", linewidth=2.5, color="#D62828")
plt.title("Global Average Temperature Change (1992‚Äì2022)")
plt.xlabel("Year")
plt.ylabel("Mean Temperature Change (¬∞C)")
plt.tight_layout()
plt.show()
```

# üß≠ Explanation

This line chart shows the global average temperature trend from 1992 to 2022.
We created this visual to observe how the mean temperature across all countries has changed over time.
It helps identify whether the world is warming or cooling by showing the overall direction of temperature change.
A steady upward trend in the line indicates a global rise in temperature, highlighting the impact of climate change over the past three decades.

```python
#  Top 10 Countries with Highest Average Temp Change

climate["MeanChange"] = climate_years.mean(axis=1)
top10 = climate.nlargest(10, "MeanChange")

plt.figure()
sns.barplot(data=top10, y="Country", x="MeanChange", palette="flare")
plt.title("Top 10 Countries with Highest Average Temperature Change")
plt.xlabel("Average Temperature Change (¬∞C)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

 # üå°Ô∏è Explanation

This bar chart displays the top 10 countries with the highest average temperature change across the years.
We created this visual to identify the regions most affected by climate change and to compare the intensity of warming between countries.
It highlights how temperature rise is not uniform across the globe, with certain nations experiencing much sharper increases in average temperature than others.

```python
#  Distribution of Temperature Change (Histogram + KDE)
plt.figure()
sns.histplot(climate["MeanChange"], bins=20, kde=True, color="#1D3557")
plt.title("Distribution of Average Temperature Change Across Countries")
plt.xlabel("Average Temperature Change (¬∞C)")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

# üßä Explanation:
This Histogram with KDE (Kernel Density Estimate) visual was created to analyze the distribution of average temperature change across different countries. It helps identify how global temperature variations are spread ‚Äî whether most countries experienced mild increases or if there are significant extremes. The KDE curve smooths the data, giving a clearer picture of overall climate patterns and highlighting global warming trends effectively.

```python
#  Before vs After 2000 Comparison (Bar Chart)

before_cols = [col for col in year_cols if int(col[1:]) <= 2000]
after_cols = [col for col in year_cols if int(col[1:]) > 2000]

climate["Before2000"] = climate[before_cols].mean(axis=1)
climate["After2000"] = climate[after_cols].mean(axis=1)
comparison = climate[["Country", "Before2000", "After2000"]].melt(id_vars="Country",
                                                             var_name="Period",
                                                             value_name="AvgTempChange")
top10_compare = comparison[comparison["Country"].isin(top10["Country"])]

plt.figure()
sns.barplot(data=top10_compare, x="AvgTempChange", y="Country", hue="Period", palette="coolwarm")
plt.title("Temperature Change Comparison (Before vs After 2000)")
plt.xlabel("Average Temperature Change (¬∞C)")
plt.ylabel("")
plt.legend(title="Period")
plt.tight_layout()
plt.show()
```

# üå°Ô∏è Explanation:
This **Bar Chart** compares the **average temperature change before and after the year 2000** for the top 10 countries. It visually highlights how temperature patterns have shifted over time, showing which regions experienced the most significant rise in temperature after 2000. By contrasting the two periods, this chart effectively illustrates the accelerating impact of climate change in the 21st century.


```python
#  Correlation Heatmap Between Years

corr = climate_years.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, cmap="coolwarm", center=0, cbar_kws={'label': 'Correlation'})
plt.title("Correlation Between Temperature Change Over the Years")
plt.tight_layout()
plt.show()
```

# üî• Explanation:
This **Correlation Heatmap** visualizes the **relationship between temperature changes across different years**. Each cell represents how strongly temperature values in one year are correlated with another. High correlations (shown in warmer colors) indicate consistent global warming trends over time, while lower correlations reveal periods of irregular temperature variation. This heatmap helps identify long-term climate patterns and temporal dependencies in global temperature change data.


```python
#  Yearly Variance Across Countries (Boxplot)

melted = climate.melt(id_vars=["Country"], value_vars=year_cols, var_name="Year", value_name="TempChange")
melted["Year"] = melted["Year"].str[1:].astype(int)

plt.figure(figsize=(14, 7))
sns.boxplot(data=melted, x="Year", y="TempChange", color="#457B9D", fliersize=2)
plt.title("Temperature Change Distribution by Year (1961‚Äì2022)")
plt.xlabel("Year")
plt.ylabel("Temperature Change (¬∞C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

üåç Explanation:
This **Boxplot** illustrates the **distribution and variability of temperature change across countries for each year (1992‚Äì2022)**. It shows the median, spread, and outliers of temperature change, helping to identify years with unusual climate fluctuations. Wider boxes or more outliers indicate greater inconsistency among countries, while narrower boxes suggest more uniform global temperature trends. This visual effectively highlights how climate variability has evolved over time.


```python
import plotly.express as px

year_col = "F2022"

# Prepare a smaller dataframe with relevant info
map_climate = climate[["ISO3", "Country", year_col]].copy()
map_climate.rename(columns={year_col: "TempChange"}, inplace=True)

# Plot the Choropleth Map using Plotly Express
fig = px.choropleth(map_climate,
                    locations="ISO3",
                    color="TempChange",
                    hover_name="Country",
                    color_continuous_scale="RdBu", # Changed from 'coolwarm' to 'RdBu'
                    title="Global Temperature Change by Country (2022)\nBased on Climate Change Indicators Dataset")

fig.update_layout(
    title_font_size=16,
    title_font_color="#D84315",
    margin={"r":0,"t":50,"l":0,"b":0}
)

fig.show()
```
